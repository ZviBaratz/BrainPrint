```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from brainprint.recon_all.results import load_results
```

```python
results = load_results(questionnaire_only=False)
```

```python
results.context["Age"] = np.round(
    (
        pd.to_datetime(results.context["Session Time"].dt.date) - results.context["Date of Birth"]
    ).dt.days / 365.2425
).astype(int)
```

```python
singles = results.context.groupby("Subject ID").last()
```

```python
g = sns.displot(singles, x="Age", hue="Sex", legend=True, aspect=2)
plt.title("Age Distribution by Sex", fontsize=14)
sns.move_legend(g, "upper right")
plt.tight_layout()
```
